```python
# importing the required modules
import numpy as np #for linear regression
import pandas as pd #for processing, csv file reading
import matplotlib.pyplot as plt #for plotting / data visualization
import seaborn as sns #for plotting
%matplotlib inline

# ml for regression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# control dispaly (optional)
pd.options.display.max_columns = 50
sns.set(style="darkgrid", rc={'axes.facecolor' : '#CAF1DE'})
# ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# load / read the dataset
df = pd.read_csv('heart_disease.csv')
df.head()
```

```python
# show people with heart diseases 
# use the column target to get the numbers
df.target.value_counts()
```

```python
# show percentage of male and female
count_female = len(df[df.sex == 0])
count_male = len(df[df.sex == 1])

# show the percentage of female patients 
print(f"Percentage os female patients: {round(count_female/(len(df.sex)) * 100 , 2)}%")
print(f"Percentage os male patients: {round(count_male/(len(df.sex)) * 100 , 2)}%")

count_male
```

```python
# finding the mean of the data 
df.groupby('target').mean()
```

```python
# plotting a bar graph showing heart disease frequency of age
pd.crosstab(df.age, df.target).plot(kind='bar', figsize=(20,6))
plt.title('Heart diseases Frequency for Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
