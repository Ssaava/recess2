```python
# importing the required libraries
import pandas as pd
```

```python
# 1.	Write a Pandas program to read the given csv file.

dataSet = pd.read_csv("AfricaCupofNationsMatches.csv")

# 2. reading the first 7 rows of the dataset
first_7_rows = dataSet.head(7)
first_7_rows
```

```python
# 3. Select the desired columns
selected_columns = dataSet[['HomeTeam', 'AwayTeam', 'HomeTeamGoals', 'AwayTeamGoals']]

# Print the selected columns
selected_columns
```

```python
# 4. selecting rows where egypt appears
egypt_rows = dataSet[(dataSet['HomeTeam'] == 'Egypt') | (dataSet['AwayTeam'] == 'Egypt') ]
egypt_rows
```

```python
# 5. counting numbers of rows and columns
num_rows = dataSet.shape[0]
num_cols = dataSet.shape[1]
print("Number of rows:", num_rows)
print("Number of columns:", num_cols)
```

```python
# 6. selecting rows where attendence is missing 
missing_attendance_rows = dataSet[dataSet['Attendance'].isnull()]
missing_attendance_rows
```

```python
# 7. Selecting the rows where 'HomeTeamGoals' are between 3 and 6 inclusive:
goals_rows = dataSet[(dataSet['HomeTeamGoals'] >= 3) & (dataSet['HomeTeamGoals'] <= 6)]
goals_rows
```

```python
# 8. Changing the "AwayTeamGoals" in the 3rd row to 10:
dataSet.loc[2, 'AwayTeamGoals'] = 10
dataSet
```

```python
# 9. Sorting the DataFrame by 'HomeTeam' in ascending order, then by 'HomeTeamScores' in descending order:
sorted_df = dataSet.sort_values(by=['HomeTeam', 'HomeTeamGoals'], ascending=[True, False])
sorted_df
```

```python
# 10. Getting a list of DataFrame column headers:
column_headers = dataSet.columns.tolist()
print(column_headers)
```

```python
# 11. Appending a column of your choice to the DataFrame:
# creating a list of values to be added to the New Column
count = len(dataSet)
new_column = ["Column Values"] * count

# adding the values to the new column
if len(new_column) == len(dataSet):
    dataSet['NewColumn'] = new_column
else:
    print("Length of new_column does not match the DataFrame length.")
dataSet
```

```python
# 12. Adding 2 rows to the DataFrame:
# Specify the headers
headers = ['Year', 'Date', 'Time', 'HomeTeam', 'AwayTeam', 'HomeTeamGoals', 'AwayTeamGoals', 'Stage',
           'SpecialWinConditions', 'Stadium', 'City', 'Attendance', 'NewColumn']

# Specify the new rows and store the values in a dictionary
new_row1 = {'Year': 2022, 'Date': '2022-01-01', 'Time': '15:00', 'HomeTeam': 'Team1', 'AwayTeam': 'Team2',
             'HomeTeamGoals': 2, 'AwayTeamGoals': 3, 'Stage': 'Group Stage', 'SpecialWinConditions': '',
             'Stadium': 'Stadium1', 'City': 'City1', 'Attendance': 50000, 'NewColumn': 1}
new_row2 = {'Year': 2022, 'Date': '2022-01-02', 'Time': '18:00', 'HomeTeam': 'Team3', 'AwayTeam': 'Team4',
             'HomeTeamGoals': 4, 'AwayTeamGoals': 1, 'Stage': 'Group Stage', 'SpecialWinConditions': '',
             'Stadium': 'Stadium2', 'City': 'City2', 'Attendance': 45000, 'NewColumn': 2}

# convert the dictionary to a dataframe and assing the index values
df1 = pd.DataFrame(new_row1, index=[count])
df2 = pd.DataFrame(new_row2, index=[count + 1])
#

# Concat dataframes
#
new_look = pd.concat([dataSet, df1])
dataSet = pd.concat([new_look, df2])
dataSet
```

```python
# 13. Changing 'Uganda' to 'China' in the 'AwayTeam' column:

dataSet.loc[dataSet['AwayTeam'] == 'Egypt', 'AwayTeam'] = 'China'
dataSet

```

```python
# 14. Resetting the index of the DataFrame:
dataSet.reset_index(drop=True, inplace=True)
dataSet
```

```python
# 15. Checking whether the 'Stadium' column is present in the DataFrame:
if 'Stadium' in dataSet.columns:
    print("The 'Stadium' column is present.")
else:
    print("The 'Stadium' column is not present.")
```

```python
# 16. Converting the datatype of the 'AwayTeamGoals' column from int to float:
dataSet['AwayTeamGoals'] = dataSet['AwayTeamGoals'].astype(float)
dataSet
```

```python
# 17. Removing the last 10 rows from the DataFrame:
dataSet = dataSet[:-10]
dataSet
```

```python
# 18. Iterating over rows in the DataFrame:
for index, row in dataSet.iterrows():
    print(row)
```

```python
# 19. Changing the order of DataFrame columns:
column_order = ['AwayTeamGoals', 'HomeTeam', 'AwayTeam', 'Attendance',  'HomeTeamGoals',  'Stadium', 'City',]
dataSet = dataSet[column_order]
dataSet
```
