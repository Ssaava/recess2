```python
# Mine csv
df2 = pd.read_csv("mine.csv")
df2
```

```python
# using fillna() method to fill in missing values with a 0
filled_nulls = df2.fillna(0)
filled_nulls
```

```python
# finding duplicates in the data using the duplicated() function
dupplicates = filled_nulls[filled_nulls.duplicated()]
dupplicates
```

```python
# removing duplicated using the drop_duplicates() function
dropped_dupplicates = filled_nulls.drop_duplicates()
dropped_dupplicates
```

```python
# function to identify wrong formats of the data
incorrect_formats = dropped_dupplicates.isna().sum()

print("Rows with incorrect numeric format:")
print(incorrect_formats)
```

```python
# dropping incorrect_formats
dropped_incorrect_formats = incorrect_formats.dropna()
dropped_incorrect_formats
```

```python
# Correlation in the data

# Calculate the correlation matrix
correlation_matrix = dropped_dupplicates.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
